```python
import altair as alt
from vega_datasets import data

alt.Chart(df,title="Age in each Profession according to Gender").mark_bar(size=2).encode(
   alt.X('Age',
        scale=alt.Scale(domain=(5,1000))
    ),
    y='Profession',
    color=alt.Color('Gender', scale=alt.Scale(scheme='dark2'))
).configure_axis(
    grid=False
)
 
```

Čiarový graf zobrazujúci vek zákazníkov v každej profesii rozdelený podľa pohlavia s niektorými z možných úprav knižnice altair. 

```python
import altair as alt
from vega_datasets import data

def pink_marks():                                        #definovanie vlastného štýlu
    return {
        'config': {
            'view': {
                'height': 300,
                'width': 500,
            },
            'mark': {
                'color': 'magenta',
                'fill': 'cyan'
            }
        }
    }

# zaregistrovanie vlastnej témy pod jej názvom
alt.themes.register('pink_marks', pink_marks)

# povolenie novej témy
alt.themes.enable('pink_marks')

# vykreslenie grafu s novou témou
alt.Chart(df).mark_point().encode(
    x='Age',
    y='Profession'
)
```

Definovania, zaregistrovanie a použitie vlastného štýlu v rámci knižnice Altair. V tomto prípade sme nastavili obrysovú farbu bodov na ružovú a výplň na tyrkysovú. 

```python
alt.themes.enable('default')


alt.Chart(df).mark_bar().encode(
    alt.X("Age"),
    y='count()',
)
```
